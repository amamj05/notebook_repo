## url에 헤더주기
- user-agent (아래는 가라헤더)
  > headers = {'User-Agent':('Mozilla/5.0(Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
                  '(KHTML, like Gecko) Chrome/68.0.3440.75 Safari/537.36'  }

```python
myheaders = {'User-Agent':('Mozilla/5.0(Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
                  '(KHTML, like Gecko) Chrome/68.0.3440.75 Safari/537.36')}

url = "https://www.melon.com/chart/index.htm"
html = requests.get(url, headers=myheaders).text
html
```

```python
# 수민님

from selenium import webdriver
from bs4 import BeautifulSoup
import time

wd = webdriver.Chrome()

image_urls = []
wd.get("https://pixabay.com/ko/images/search/강아지/")

time.sleep(5)

soup = BeautifulSoup(wd.page_source, 'html.parser')
image_divs = soup.find_all('div', class_='cell--B7yKd')

for i in range(min(100, len(image_divs))):
    img_tag = image_divs[i].find('img')
    
    if img_tag and 'src' in img_tag.attrs:
        img_url = img_tag['src']
        image_urls.append(img_url)

print(image_urls)
```
